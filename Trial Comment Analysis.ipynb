```python
%matplotlib inline
import nltk
from nltk.corpus import stopwords
#from nltk.stem import WordNetLemmatizer
stop_words = set(stopwords.words('english'))
from collections import Counter
#from nltk.probability import FreqDist
import pandas as pd
```

```python
df = pd.read_csv('test_comments1.csv')
df

#overview of 1st 20 commentators + comment & metadata of testfile dataset
```

```python
#retrieve comments from table and format them as a list to prep for cleanup and analysis
list_vice_comments = (list(df['Comment']))
text = ' '.join(list_vice_comments)
```

```python
#regular tokenization and preprocessing: 
def clean_text(data):
    data = data.lower()
    punctuation_and_numericals = '!@#$%^&*()_-+={}[]:;"\'|<>,.?/~`1234567890'
    return ''.join(character for character in data
                 if character not in punctuation_and_numericals)

all_tokens = clean_text(text).split()
topic_tokens = [w for w in all_tokens if not w in stop_words]
freq = Counter(topic_tokens)
freq.most_common(100)
```
